```python
 import math

def f_bisection(x):
    return math.sin(x) - x

def bisection(a, b, tol):
    if f_bisection(a) * f_bisection(b) >= 0:
        print("Metode gagal: f(a) dan f(b) tidak memiliki tanda berlawanan.")
        return None

    while (b - a) / 2.0 > tol:
        c = (a + b) / 2.0
        if f_bisection(c) == 0:
            return c
        elif f_bisection(a) * f_bisection(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2.0

root_bisection = bisection(0, 1, 0.01)
print(f"Akar dengan Bisection: {root_bisection}")
```

```python
def f_newton(x):
    return x*3 - 3*x*2 + 2*x

def df_newton(x):
    return 3*x**2 - 6*x + 2

def newton_raphson(x0, tol, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f_newton(x)
        dfx = df_newton(x)

        print(f"Iterasi {i+1}: x = {x}, f(x) = {fx}, f'(x) = {dfx}")

        if dfx == 0:
            print("Turunan bernilai nol. Metode gagal.")
            return None

        x_new = x - fx / dfx

        if abs(x_new - x) < tol:
            return x_new
        x = x_new

    print("Maksimum iterasi tercapai. Metode gagal konvergen.")
    return None

akar_newton = newton_raphson(1.5, 0.01)
print(f"Akar dengan Newton-Raphson: {akar_newton}")
```
